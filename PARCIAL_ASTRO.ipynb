PUNTO C

FORMULAS UTILIZADAS SACADAS DEL LIBRO GUIA

```python
# WGS84   #Parametros que necesito
a = 6378137             # el semieje mayor
f = 1 / 298.257       #  el achatamiento
```

```python
def Geo(latitud, altura):
    # Convertimos a radianes (las funciones trigonometricas usan radianes)
    latitudRad = np.radians(latitud)

    latitudGeo= np.arctan((1-f)**2 * np.tan(latitudRad)) #pasamos a latitud geocentrica
    #calculamos la excentridad del elipsoide
    e= np.sqrt(f *(2-f))

    Ro = a * np.sqrt((1 - e**2 * (2 - e**2) * np.sin(latitudRad)**2) / #Distancia radial
                     (1 - e**2 * np.sin(latitudRad)**2))
    altura= altura + Ro #le sumamos la altura inicial

    return latitudGeo, altura
```

```python
# Datos de las estaciones, hacemos una tupla con 4 elementos cada una
estaciones = [("La Serena, Chile", -29.9, -71.25, 29),("Dakar, Senegal", 14.7, -17.45, 24),("Canarias, Espana", 28.3, -16.5, 2390),("Munich, Alemania", 48.14, 11.58, 520),
    ("Sutherland, Sudafrica", -32.38, 20.81, 1450),
    ("Siding Spring, Australia", -31.27, 149.07, 1165)
]

```

```python
for nombre, lat, lon, h in estaciones: #hacemos un for que pueda iterar en cada elemento en la lista estaciones
    lat_geo, Ro = Geo(lat,h) #llamamos a la funcion


    print(nombre, "\n") # Imprime el nombre de la estacion
    print(f"  Latitud Geocéntrica = {np.degrees(lat_geo):.3f}°") #convertimos a grados
    print(f"  Longitud = {lon:.3f}°")  # mantenemos la longitud original
    print(f"  Distancia Radial = {Ro:.3f} m\n") #.3f redondea a 3 decimales
```

PUNTO D

```python

 #######ESTE CODIGO FUE USADO EN LA TAREA ENVIADA ANTERIORMENTE EN EL CALCULO DE DISTANCIAS##########


def distancia(lat1, lon1, lat2, lon2):
    R = 6371.0

    lat1_rad = np.radians(lat1)
    lat2_rad = np.radians(lat2)
    lon1_rad = np.radians(lon1)
    lon2_rad = np.radians(lon2)


    if (lon1 * lon2) > 0: #si estan en el mismo hemisferio en longitud resto

        AnguloC = abs(lon1 - lon2)
    elif (abs(lon1) + abs(lon2)) <= 180: # si estan en hemsferios diferentes sin cruzar el antimeridiano sumo

        AnguloC = abs(lon1) + abs(lon2)
    else:
        AnguloC = 360 - (abs(lon1) + abs(lon2)) #si cruzo el antimeridiano


    AnguloCRAD = np.radians(AnguloC)

    # Teorema del coseno esferico
    C = np.sin(lat1_rad) * np.sin(lat2_rad) + np.cos(lat1_rad) * np.cos(lat2_rad) * np.cos(AnguloCRAD)
    C = np.clip(C, -1, 1)  # recorta cualquier valor fuera del rango permitido.

    distancia = np.arccos(C) * R

    return distancia
```

```python

n = len(estaciones) #longitud de la lista estaciones
letras = [chr(65 + i) for i in range(n)]         #  chr() convierte un numero (codigo ASCII) en su caracter adecuado,   Por ejemplo: chr(65) = A chr(66) = B y recorre en n aumetando 65+0,65+1, etc..
D = np.zeros((n, n)) #Crea una matriz cuadrada de ceros de tamaño n × n , esta matriz sirve para guardar las distancias entre las estaciones
for i in range(n): # recorre todos los índices de las estaciones
    for j in range(n): # representa la estación de destino, la columna de la matriz.
    #para cada estación de origen i, estamos revisando todas las estaciones destino j.
        if i == j: #Si la estacion de origen y destino son la misma (i == j), la distancia es 0 km. llena la diagonal de ceros
            D[i, j] = 0.0
        else:
            lat1, lon1 = estaciones[i][1], estaciones[i][2]  #la latitud de la estacion i y la longitud de la estacion i
            #i = 0 (La Serena) = lat1 = -29.9, lon1 = -71.25
            lat2, lon2 = estaciones[j][1], estaciones[j][2]
            #j = 1 (Dakar) = lat2 = 14.7, lon2 = -17.45
            D[i, j] = distancia(lat1, lon1, lat2, lon2) #llamo a la funcion

df = pd.DataFrame(D, index=letras, columns=letras) #convierto en un dataframe(tabla con la matriz d, filas=LETRAS y columnas= Letras)
pd.set_option("display.precision", 2) #Esto le dice a pandas cuantos decimales mostrar al imprimir números
print("Matriz de distancias (km):\n")
print(df.round(2)) #REDONDEO

```

  PUNTO E

CONDICIONES SI CRUZA EL ECUADOR O EL MERIDIANO

SI ES FALSE, NO CRUZA SI ES TRUE, SI CRUZA

```python

resultados = []
 ######MISMA LOGICA CODIGO ANTERIOR###########
# Recorremos todas las estaciones posibles (origen - destino)
for i in range(n):
    for j in range(n):
        if i != j:
            nom1, lat1, lon1, _ = estaciones[i]  # el _ ignora la altitud  #Desempaqueta la tupla y le asigna valores
            nom2, lat2, lon2, _ = estaciones[j]

            # Llamamos a la funcion distancia
            dist = distancia(lat1, lon1, lat2, lon2)

            # Revisamos si cruza el ecuador (una latitud positiva y la otra negativa)
            cruza_ecuador = (lat1 * lat2) < 0

            # Revisamos si cruza el meridiano de Greenwich (una longitud positiva y la otra negativa)
            cruza_greenwich = (lon1 * lon2) < 0

            # Guardamos los datos en un diccionario
            resultados.append({
                "Origen": nom1,
                "Destino": nom2,
                "Distancia_km": round(dist, 2),
                "Cruza_Ecuador": cruza_ecuador,
                "Cruza_Greenwich": cruza_greenwich
            })

df1 = pd.DataFrame(resultados) #Convertimos la lista de resultados a un DataFrame
pd.set_option("display.precision", 2) # Le decimos a pandas que muestre solo 2 decimales
print("Trayectos entre estaciones:\n")
print(df1)
```

```python


lat_origen, lon_origen = 28.100, -16.500   # Canarias
lat_destino, lon_destino = -31.275, 149.066  # Siding Spring

# Convertir grados a radianes para trigonometria
lat1 = np.radians(lat_origen)
lon1 = np.radians(lon_origen)
lat2 = np.radians(lat_destino)
lon2 = np.radians(lon_destino)

# Diferencia de longitudes
d_lon = lon2 - lon1

# Rumbo inicial usando formula de triqngulo esferico
y1 = np.cos(lat2) * np.sin(d_lon) # componente este-oeste
x1 = np.cos(lat1)*np.sin(lat2) - np.sin(lat1)*np.cos(lat2)*np.cos(d_lon) # componente norte-sur
alpha1 = np.arctan2(y1, x1)  # en radianes

# Rumbo final usando fórmula inversa
y2 = -np.cos(lat1) * np.sin(d_lon)
x2 = np.cos(lat2)*np.sin(lat1) - np.sin(lat2)*np.cos(lat1)*np.cos(d_lon)
alpha2 = np.arctan2(y2, x2) + np.pi  # sumamos pi para invertir el rumbo

# Convertir a grados y normalizar 0-360 grados,  ## wrap360 convierte cualquier angulo en grados a un valor entre 0 y 360 grados,  atan2 puede devolver angulos negativos
def wrap360(deg):
    d = deg % 360  # toma el residuo al dividir entre 360
    return d if d >= 0 else d + 360  # asegura que el angulo sea positivo

alpha1 = wrap360(np.degrees(alpha1)) # Rumbo inicial en grados
alpha2 = wrap360(np.degrees(alpha2)) # Rumbo final en grados

print(f"Rumbo inicial: {alpha1:.2f}°")
print(f"Rumbo final: {alpha2:.2f}°")
```
